```python
# 결측치 확인하기
idx_missing=np.isnan(df_contin_np)
print(idx_missing)
```

```python
print('전체 결측치의 개수:', np.isnan(df_contin_np).sum())
print('행별 결측치의 개수:', np.isnan(df_contin_np).sum(1))
print('열별 결측치의 개수:', np.isnan(df_contin_np).sum(0))
```

```python
df_contin_np.shape
```

```python
# sklearn.impute에서 모두 불러오기
from sklearn.impute import *
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
```

```python
?SimpleImputer
```

```python
# 평균으로 채워넣기
imp_mean = SimpleImputer(strategy='mean', missing_values=np.nan) 
# 1. 모형 선언,  결측치는 평균으로 채워넣고 결측치는 np.nan 값을 가지고
imp_mean.fit(df_contin_np)
# 2. 각 열별 평균을 EQ_contin_np 행렬을 기준으로 계산
```

```python
# 각 열의 fit된 평균값 출력
imp_mean.statistics_
```

```python
# 3. transform : statistics_에 저장된 값으로 결측치를 채워 넣겠다. 
print(imp_mean.transform(df_contin_np))
```

```python
# median으로 채워넣기
imp_median = SimpleImputer(strategy='median', missing_values=np.nan) # 선언, median으로 결측치를 채워 넣겠다
imp_median.fit(df_contin_np) # median 값을 학습
```

```python
imp_median.statistics_
```

```python
print(imp_median.transform(df_contin_np))
```
