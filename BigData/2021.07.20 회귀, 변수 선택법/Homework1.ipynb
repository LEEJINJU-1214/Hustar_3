```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_boston

boston = load_boston()
print(boston.DESCR)
```

```python
dfX = pd.DataFrame(boston.data, columns=boston.feature_names)
dfy = pd.DataFrame(boston.target, columns=["MEDV"])
df = pd.concat([dfX, dfy], axis=1)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Q1. 원하는 변수를 2개 이상을 가지고 'price' 예측하는 모델 만들기

#### sckit-learn 활용, R^2/MSE 계산

```python
# correlation 시각화
plt.figure(figsize=(20,10))

sns.heatmap(data = df.corr(), annot=True, 
            fmt = '.2f', linewidths=.5, cmap='Blues')

#각 독립변수간의 상관관계가 존재함
#종속변수와 0.45이상의 상관관계를 가진 변수 5개
#후진선택법으로 5개의 변수를 선택함
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
bs = SFS(LinearRegression(),
           k_features=5, forward=False, floating=False,scoring = 'r2', cv = 0)

bs.fit(dfX,dfy)
print(f'후진제거법시 R^2: {bs.k_score_:.4f}')
print('후진제거법시 Feature Index: ', bs.k_feature_idx_)
```

```python
dfx_bs = dfX.iloc[:,[4, 5, 7, 10, 12]]
```

```python
plt.figure(figsize=(20,10))

sns.heatmap(data = dfx_bs.corr(), annot=True, 
            fmt = '.2f', linewidths=.5, cmap='Blues')

#선택된 변수간의 상관관계가 존재함
```

```python
X_train, X_test, y_train, y_test = train_test_split(dfx_bs,
                                                    dfy,
                                                    test_size=0.2,
                                                    random_state=5)
#0.8로 split
```

```python
# train set만을 이용해서 모형 학습
rl= LinearRegression()
rl.fit(X_train, y_train)
# train set에서의 성능
y_train_hat = rl.predict(X_train)
print('train셋 MSE: ', mean_squared_error(y_train,y_train_hat))
print('train셋 R^2: ', rl.score(X_train, y_train))
# test set에서의 성능
y_test_hat = rl.predict(X_test)
print('*' *30)
print('test셋 MSE:', mean_squared_error(y_test,y_test_hat))
print('test셋 R^2: ', rl.score(X_test, y_test))
```
